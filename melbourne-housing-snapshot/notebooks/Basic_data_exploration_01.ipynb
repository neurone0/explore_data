#We will define a function for optimizing max leaf nodes to avoid overfitting and underfitting.


```python
def get_mae(max_leaf_nodes,train_X,train_y,val_X,val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state= 0)
    model.fit(train_X,train_y)
    predict_value =model.predict(val_X)
    mae = mean_absolute_error(val_y,predict_value)
    return mae
    
```

```python
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]
```

```python
for maxnode in candidate_max_leaf_nodes:
    mymei = round(get_mae(maxnode,train_X,train_y,val_X,val_y),2)
    print(mymei)
```

```python
#annother sol using dict comprehensive
scores = {leaf_size: get_mae(leaf_size,train_X,train_y,val_X,val_y) for leaf_size in candidate_max_leaf_nodes}
print(scores)
```

```python
best_tree_size = min(scores, key=scores.get)
```

```python
best_tree_size
```

```python
# Fill in argument to make optimal size and uncomment
final_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=1)

# fit the final model and uncomment the next two lines
final_model.fit(X, y)
```

```python
final_pred = final_model.predict(val_X)
final_mae = mean_absolute_error(final_pred,val_y)
print(final_mae)
```
